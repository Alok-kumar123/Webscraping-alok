```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url='https://stores.lenskart.com/'
data=requests.get(url).text
soup=BeautifulSoup(data,'html.parser')
d=soup.find('div',{'class':'store-list'})
```

```python
Data=pd.DataFrame(columns=["Name","Address","Timings","Contact","Latitude&Longitude"])
```

```python
for k in d.find_all('a'):
    
    name=str(k.contents[0])
    print(name)
    ur=k.get('href')
    
    Data1=requests.get(ur).text
    
    Soup1=BeautifulSoup(Data1,'html.parser')
    Daata=Soup1.find_all('div',{'class':'store-info'})
    
    add=''
    for i in Daata:
        for j in np.arange(0,6,2):
            add=add+str(i.find('span',{'class':'address'}).contents[j]) 
     
        name=i.find('span').contents[0]
         
    
        contact=i.find('a',{'class':'tel'}).contents[1]
        timing=i.find('span',{'class':'store-timing'}).contents[1] 
        n=i.find('a',{'class':'store-details'}).get('href')
        lat_long = n[n.find('&query=')+ len('&query='):n.find('&query_place')]
        Data=Data.append({"Name":name,"Address":add,"Timings":timing,"Contact":contact,"Latitude&Longitude":lat_long},ignore_index=True)
        add=''
```
